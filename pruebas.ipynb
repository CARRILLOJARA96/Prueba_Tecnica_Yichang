```python
BD_Atun=pd.read_csv("BD_Evaluación_DS/Categorías/BD_Atún.csv", sep=';',encoding='latin-1')
BD_Cereales=pd.read_csv("BD_Evaluación_DS/Categorías/BD_Cereales.csv", sep=';',encoding='latin-1')
BD_Snacks=pd.read_csv("BD_Evaluación_DS/Categorías/BD_Snacks.csv", sep=';',encoding='latin-1')
```

```python


def etl_procesar_datos(df):

    # Utilizar el método rename con str.replace para quitar los espacios de los nombres de las columnas
    df.rename(columns=lambda x: x.replace(' ', ''), inplace=True)
    
    # Verificar si la columna 'Unnamed:9' existe en el DataFrame antes de eliminarla
    if 'Unnamed:9' in df.columns:
        df = df.drop(columns='Unnamed:9')
    
    # Reemplaza comas, espacios y guiones por una cadena vacía en todo el DataFrame usando una expresión regular
    df = df.replace(r'[, -]', '', regex=True)

    # Reemplazamos los vacíos por NULO
    df.replace('', pd.NA, inplace=True)

    #cambiamos los tipos de datos de las columnas en todo el dataframe
    df['Día'] = pd.to_datetime(df['Día'], format='%d/%m/%Y')
    df['Mes'] = df['Mes'].astype(str)
    df['Retail'] = df['Retail'].astype(str)
    df['Zona'] = df['Zona'].astype(str)
    df['Venta_Unidades'] = pd.to_numeric(df['Venta_Unidades'], errors='coerce')
    df['Venta_Soles'] = pd.to_numeric(df['Venta_Soles'], errors='coerce')
    df['PVP_Unitario'] = pd.to_numeric(df['PVP_Unitario'], errors='coerce')
    df['Inventario_Unidades'] = pd.to_numeric(df['Inventario_Unidades'], errors='coerce')

    # Imputa valores faltantes con la media de cada columna
    df['Venta_Unidades'].fillna(df['Venta_Unidades'].mean(), inplace=True)
    df['Venta_Soles'].fillna(df['Venta_Soles'].mean(), inplace=True)
    df['PVP_Unitario'].fillna(df['PVP_Unitario'].mean(), inplace=True)
    df['Inventario_Unidades'].fillna(df['Inventario_Unidades'].mean(), inplace=True)

    # Devolvemos el DataFrame con el nombre correspondiente
    return df
```
